```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import sklearn as skl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
flight=pd.read_csv("EastWestAirlines.csv")
flight
```

```python
print(flight.info())
```

```python
flight1 = flight.drop(['ID#','Award?'],axis=1)
flight1.head()
```

```python
array=flight1.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X 
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=11)
dbscan.fit(X)    
```

```python
dbscan.labels_ 
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
pd.set_option("display.max_rows", None)
```

```python
cl
```

```python
flight1 = pd.concat([flight,cl],axis=1) 
flight1     
```

```python
import matplotlib.pyplot as plt
>>> plt.style.use('classic')
```

```python
flight1.plot(x="Balance",y ="Bonus_trans",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')  
```

```python
dl = dbscan.labels_ 
```

```python
skl.metrics.silhouette_score(X, dl)
```
